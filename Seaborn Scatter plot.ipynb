```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tt_df = sns.load_dataset("titanic")
tt_df
```

```python
sns.scatterplot(x = "age", y = "fare", data = tt_df)

plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = tt_df, hue = "sex")

plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = tt_df, hue = "sex",
               style = "who")

plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = tt_df, hue = "sex",
               style = "who", size = "who", sizes = (100, 300))

plt.show()
```

```python
sns.set(style = ("darkgrid"))

plt.figure(figsize = (16,9))

sns.scatterplot(x = "age", y = "fare", data = tt_df, hue = "sex",
               style = "who", size = "who", sizes = (100, 300))

plt.show()
```

```python
sns.set(style = ("darkgrid"))

plt.figure(figsize = (16,9))

sns.scatterplot(x = "who", y = "fare", data = tt_df, hue = "alive",
               style = "alive", size = "who", sizes = (100, 300))

plt.show()
```

```python
sns.set(style = ("darkgrid"))

plt.figure(figsize = (16,9))

sns.scatterplot(x = "who", y = "fare", data = tt_df, hue = "alive",
               style = "alive", size = "who", sizes = (100, 300),
               palette = "gist_heat", alpha = 0.8)

plt.show()
```

```python
plt.figure(figsize = (16,9))

sns.scatterplot(x = "age", y = "fare", data = tt_df)
sns.lineplot(x = "age", y = "fare", data = tt_df)
sns.barplot(x = "age", y = "fare", data = tt_df)


plt.show()
```
