```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as norm
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.displot(tips_df["total_bill"])

plt.show()
```

```python
sns.displot(tips_df["total_bill"], kde = True)

plt.show()
```

```python
plt.figure(figsize = (16,9))

sns.set()
sns.displot(tips_df["total_bill"], kde = True, rug = True, color = "m",
           label = "Total bill")

plt.title("Histogram of Total bill")
plt.legend()
plt.show()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

plt.figure(figsize = (16,9))

sns.set()
sns.displot(tips_df["total_bill"], bins = bins, kde = True, rug = True, color = "m",
           label = "Total bill")

plt.xticks(bins)

plt.title("Histogram of Total bill")
plt.legend()
plt.show()
```
