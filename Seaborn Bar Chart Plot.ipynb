```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)

plt.show()
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)

plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df)

plt.show()
```

```python
order = ["Sun", "Thur", "Fri", "Sat"]

sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df, order = order)

plt.show()
```

```python
hue_order = ["Female", "Male"]

sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df, hue_order = hue_order)

plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df, estimator = np.max)

plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df, ci = 20)

plt.show()
```

```python
sns.barplot(y = "day", x = "total_bill", hue = "sex",
           data = tips_df)

plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df, palette = "hot")

plt.show()
```

```python
kwargs = {"alpha":0.9, "linestyle":"-", "linewidth":2, "edgecolor":"k"}

sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df, palette = "hot", **kwargs)

plt.show()
```

```python
kwargs = {"alpha":0.9, "linestyle":"-", "linewidth":2, "edgecolor":"k"}

ax = sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df, palette = "hot", **kwargs)

ax.set(title = "Total Bill Bar Chart",
      xlabel = "Days",
      ylabel = "Total Bills")

plt.show()
```

```python
sns.set(style = "darkgrid")

plt.figure(figsize = (12,7))

kwargs = {"alpha":1, "linestyle":"-", "linewidth":2, "edgecolor":"k"}

ax = sns.barplot(x = "day", y = "total_bill", hue = "sex",
           data = tips_df, palette = "hot", **kwargs, capsize = 0.3)

plt.title("Bar Chart of Days and Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 20)
plt.ylabel("Total Bill", fontsize = 20)

plt.show()
```
