```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
GW_df = pd.read_csv("C:\\Users\\dell\\OneDrive\\Desktop\\First Project\\Who_is_responsible_for_global_warming.csv")
GW_df
```

```python
GW_df = GW_df.drop(columns = ["Country Code", "Indicator Name", "Indicator Code"], index = 1).set_index("Country Name")
GW_df.head(6)
```

```python
sns.heatmap(GW_df)

plt.show()
```

```python
sns.heatmap(GW_df, vmin = 0, vmax = 21)

plt.show()
```

```python
sns.heatmap(GW_df, vmin = 0, vmax = 21, cmap = "coolwarm")

plt.show()
```

```python
plt.figure(figsize = (13,7))

sns.heatmap(GW_df, vmin = 0, vmax = 21, cmap = "coolwarm",annot = True)

plt.show()
```

```python
plt.figure(figsize = (13,7))

annot_kws = {"fontsize":10,
            "fontstyle":"italic",
            "color":"k",
            "alpha":0.9,
            "rotation":"vertical",
            "verticalalignment":"center",
            "backgroundcolor":"w"}

sns.heatmap(GW_df, vmin = 0, vmax = 21, cmap = "coolwarm",annot = True, annot_kws = annot_kws)

plt.show()
```

```python
plt.figure(figsize = (13,7))

sns.heatmap(GW_df, vmin = 0, vmax = 21, cmap = "coolwarm",annot = True, linewidth = 3)

plt.show()
```

```python
plt.figure(figsize = (16,7))

sns.heatmap(GW_df, vmin = 0, vmax = 21, cmap = "coolwarm",annot = True, linewidth = 4, linecolor = "k")

plt.show()
```

```python
plt.figure(figsize = (16,7))

cbar_kws = {"orientation":"horizontal",
           "shrink":1,
           "extend":"min",
           "extendfrac":0.1,
           "ticks":np.arange(0,22),
           "drawedges":True}

sns.heatmap(GW_df, annot = True, cbar_kws = cbar_kws)

plt.show()
```

```python
# for customize the table according to your requirment 

country_code = ["AMERICA", "INDIA", "NEPAL", "AUSTRALIA", "CHINA", "BRAZIL", "JAPAN", "BANGLADESH", "FRANCE",]

sns.heatmap(GW_df, xticklabels = np.arange(0.22), yticklabels = country_code)

plt.show()
```

```python
plt.figure(figsize = (16,9))

ex = sns.heatmap(GW_df)

ex.set(title = "Who is reponse for global warming",
      xlabel = "CO2 emissions (matrix ton per capital) per year",
      ylabel = "Country Name")
sns.set(font_scale = 2)

plt.show()
```

```python
GW_df.corr()
```

```python
plt.figure(figsize = (18,9))

Years_code = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]

ax = sns.heatmap(GW_df.corr(), annot = True, linewidth = 3, yticklabels = Years_code)
ax.tick_params(size = 5, color = "w", labelsize = 5, labelcolor = "w")
plt.title("Who is responsible for Global Warming", fontsize = 25)

plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
                        columns = np.append(cancer_dataset['feature_names'], ['target']))
cancer_df
```

```python
plt.figure(figsize = (45,30))
ax = sns.heatmap(cancer_df.corr(), annot = True, linewidth = 3)
ax.tick_params(size = 5, color = "w", labelsize = 5, labelcolor = "w")
plt.title("heatmap of 'Breast cancer patients'", fontsize = 25)

plt.show()
```
